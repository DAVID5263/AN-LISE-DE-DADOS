```python
#Preenchendo os Valores em branco da Base De Dados
base['marital_Divorced'].fillna(base['marital_Divorced'],inplace=True)
base['marital_Married'].fillna(0,inplace=True)
base['marital_Single'].fillna('NaN',inplace=True)
base['marital_Together'].fillna('NaN',inplace=True)
base['marital_Widow'].fillna('NaN',inplace=True)
base['education_2n Cycle'].fillna('NaN',inplace=True)
base['education_Basic'].fillna('NaN',inplace=True)
base['education_Graduation'].fillna('NaN',inplace=True)
base['education_Master'].fillna('NaN',inplace=True)
base['education_PhD'].fillna(0.0,inplace=True)
base['kids'].fillna(0,inplace=True)
base['marital_status'].fillna('Outros',inplace=True)
base.info()
base['marital_Divorced'].tail(200)
```

```python
#Separando o nivel de Escolaridade
nivel1 = []
nivel2 = []
nivel3 = []
nivel4 = []
outros = []
for i in base['education_level']:
  if i == 'PhD':
    nivel1.append(i)
  elif i == 'Graduation':
    nivel2.append(i)
  elif i == 'Master':
    nivel3.append(i)
  elif i == 'Basic':
    nivel4.append(i)
  else:
    outros.append(i)
nv1_total = (len(nivel1))
nv2_total = (len(nivel2))
nv3_total = (len(nivel3))
nv4_total = (len(nivel4))
nv5_total = (len(outros))
soma_total = (len(nivel1)+len(nivel2)+len(nivel3)+len(nivel4)+len(outros))
```

```python
# Nível de escolaridade em Porcentagem
n1 = ((nv1_total/soma_total)*100)
print(f'O Nível 1 Representa {n1:.2f} %')
n2 = ((nv2_total/soma_total)*100)
print(f'o Nível 2 Representa {n2:.2f} %')
n3 = ((nv3_total/soma_total)*100)
print(f'O Nível 3 Representa {n3:.2f} %')
n4 = ((nv4_total/soma_total)*100)
print(f'O Nível 4 Representa {n4:.2f} %')
n5 = ((nv5_total/soma_total)*100)
print(f'O Nível 5 Representa {n5:.2f} %')
```

```python
#Gráfico de Barras
from matplotlib import axis
x = ['PhD','GRADUAÇÃO','MESTRADO','BÁSICO','OUTROS']
y = [n1,n2,n3,n4,n5]
dy = sorted(y,reverse=True)

# plot
fig, ax = plt.subplots()

plt.bar(sorted(x),dy, color='#21130d')
plt.title("Nível de Educação")
for i in range(0,5):
  plt.annotate('{:,.2f} %'.format(dy[i]),(i,dy[i]),
  ha="center",
  xytext=(0,5),
  textcoords="offset points",
  )
ax.legend()
ax.set(ylim=(0,60))
ax.tick_params('x', labelrotation=45,size=25)
plt.show()
```
